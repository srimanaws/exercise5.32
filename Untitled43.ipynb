```python
# Exercise 5.32 by sriman
import random
import seaborn as sns
import matplotlib.pyplot as plt

def roll_dice():
    """Simulate rolling two dice and return their sum."""
    return random.randint(1, 6) + random.randint(1, 6)

def simulate_rolls(num_rolls):
    """Simulate a certain number of dice rolls and return frequencies of sums."""
    frequencies = [0] * 11  # Index 0 won't be used (2 to 12)

    for _ in range(num_rolls):
        roll_sum = roll_dice()
        frequencies[roll_sum - 2] += 1  # Adjusting index for sums (2-12)

    return frequencies

def plot_frequencies(frequencies, rolls):
    """Plot the frequencies of the dice sums."""
    sns.barplot(x=list(range(2, 13)), y=frequencies, palette="muted", orient="v")
    plt.title(f'Frequencies for {rolls} Rolls')
    plt.xlabel('Dice Sum')
    plt.ylabel('Frequency')
    plt.show()

# Simulate and plot for different numbers of rolls
for rolls in [360, 3600, 36000]:
    freq = simulate_rolls(rolls)
    plot_frequencies(freq, rolls)
```
